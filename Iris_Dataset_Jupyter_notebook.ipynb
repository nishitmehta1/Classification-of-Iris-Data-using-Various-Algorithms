The following code is for Classifying Iris data using Naive Bayes. The training Data has been downloaded from UCI's database.

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('iris dataset/iris.csv')
data_test = pd.read_csv('iris dataset/iris_test.csv')

data = data.sample(frac=1).reset_index(drop=True)
data_test = data_test.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
x = data.iloc[:,:-1].values
y = data.iloc[:,4].values

x_test = data_test.iloc[:,:-1].values
y_test = data_test.iloc[:,4].values
```

```python
classifier = GaussianNB()
classifier.fit(x, y)
```

```python
y_pred = classifier.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
accuracy = classifier.score(x_test, y_test)
print ("Accuracy =", accuracy*100)
```
